```python
from gensim.test.utils import common_texts, get_tmpfile
from gensim.models import Word2Vec
```

```python
from TF_IDF import corpusA
```

```python
corpusA = [i.split(' ') for i in corpusA]
```

```python
corpusA
```

```python
"""
size：整数，可选
    单词向量的维数。
window：int，可选
    句子中当前词和预测词之间的最大距离。
min_count：int，可选
    忽略所有总频率低于此频率的单词。
workers：int，可选
    使用这些许多工作线程来训练模型（=使用多核机器进行更快的训练）。
sg：{0，1}，可选
    训练算法：1表示跳过语法；否则CBOW。
hs：{0，1}，可选
    如果为1，则将等级softmax用于模型训练。
    如果为0，且“负数”为非零，则将使用负数采样。
negative：int，可选
    如果> 0，将使用负采样，负的int指定多少个“噪声词”
    应绘制（通常在5-20之间）。
    如果设置为0，则不使用负采样。
ns_exponent：浮动，可选
    用于塑造负采样分布的指数。 1.0的值正好按比例采样
    在频率上，0.0对所有单词平均采样，而负值对低频单词采样更多
    比高频单词。最初的Word2Vec文件选择了流行的默认值0.75
"""

models = Word2Vec(corpusA, size=100, window=5, min_count=1, workers=4, sg=1, negative=5, ns_exponent=0.75)
```

```python
models.most_similar('最靠近')
```

```python
models.save('./skip-gram')
```
